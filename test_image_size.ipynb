```python
from PIL import Image
import torch
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np
import cv2 as cv
```

```python
path = r"/content/drive/MyDrive/test_image_LIR/1.tif"
tif_img = Image.open(path)
# tif_img.show()
plt.imshow(tif_img)
plt.axis('off')  # Turn off axis
plt.show()
```

```python
tif_img_np = np.array(tif_img)
type(tif_img_np)
tif_image_np = (tif_img_np - np.min(tif_img_np)) / (np.max(tif_img_np) - np.min(tif_img_np))
tif_image_np.shape
```

```python
# Assuming tif_img is your PIL image object
# resize_transform = transforms.Resize((424, 240))
resize_transform = transforms.Resize((240, 424))
resized_img = resize_transform(tif_img)

# Convert the resized image to a NumPy array
resized_img_np = np.array(resized_img)
print(resized_img_np.shape)

# Display the image using matplotlib
plt.imshow(resized_img)
plt.axis('off')  # Turn off axis
plt.show()
```
